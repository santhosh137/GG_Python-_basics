```
### truediv, floordiv, pow , mod

import operator
a=5
b=2
c=22
d=7
print(operator.truediv(c,d))
print(operator.floordiv(a,b))
print(operator.pow(a,b))
print(operator.mod(a,b))
```

```
### lt -< , le - <= , eq ==, 

import operator
print (operator.lt(2,3))
print (operator.le(3,2))
print (operator.eq(3,3))
```

```
### gt -> , ge- >= , ne 

import operator
print (operator.ge(3,5))
print (operator.gt(4,3))
print (operator.ne(4,5))
```

```
### setitem, getitem, delitem

##setitem (obj, pos, value) 
## getitem (obj , pos)
#delitem(object, postion)
```

```
import operator
l=[1,2,3,4,5]

print (l)

operator.setitem(l,4, 6)
print(l)
operator.getitem(l,3)
print(l)
operator.delitem(l,2)
print(l)
```

```
import operator
l=[1,2,3,4,5]

print (l)

operator.setitem(l,slice(2,4),[100,101])
print (l)

operator.setitem(l,slice(1,3),[5,6])
print (l)

```
