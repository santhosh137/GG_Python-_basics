```
### Tuples

empty_tuples=()

print (empty_tuples)

tup1=('geeks','for','geeks',5)

print (tup1)
```

```
## concatenation of tuples

tup1=('geeks','for','geeks')

tup2=(4,5,6)


print (tup1+tup2)
```

```
## nested tuples
tup1=('geeks','for','geeks')

tup2=('pyhton')

tup3=(tup1,tup2)


print (tup3)
```

```
### Repetition in tuples

tuple1=('python',)*4

print (tuple1)
```

```
## tuples are immutable

tup1=(0,1,2,3,4)
tup1[1]=-4

print (tup1)
```

```
tup1=(1,2,3)

print(len(tup1))
```
