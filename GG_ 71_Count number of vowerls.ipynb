```
n=str(input("Enter a string input: "))
print (n)
print (type(n))
count=0
for i in range(len(n)):
  if n [i] in {'a','e','i', 'o','u'}:
    count=count+1
  else:
    pass
  


print (count)
```

```
def vowel_count(str):
  count=0
  vowel=set("aeiouAEIOU")
  for alphabet in str:
    if alphabet in vowel:
      count=count+1
  
  print ("No of vowels :", count)


vowel_count("Hello World , Once again")
```
