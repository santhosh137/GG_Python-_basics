```
### index an count functions

### index function returns the index of the first element

## count returns the no of occurence of lements in the list

l=[1,1,2,2,2,3,4,7]

print (" First occurenec of 2 is" ,l.index(2,1,8))

print ("The number of occurenecs of 2 is ",l.count(2))
```

```
### Insert and remove functions

l=[1,2,3,4,4,5]

l.insert(3,2)

## here the 2 is inserted at the index 3

print (l)

l.remove(5)

### using remove to remove the element


print (l)
```
