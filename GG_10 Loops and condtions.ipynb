```
# while  loop

count=0
while count<3:
  print ("Hello world")
  count=count+1

  
```

```
## for loops

l=["This ", "is", "Geeks for Geeks"]
for i in l:
  print (i)

lp=[i for i in l]
print(lp)
```

```
### nested for  loops 

l=["This ", "is", "Geeks for Geeks"]

for word in l:
  for char in word:
    print(char)


lp=[char for char in word for word in l]

print (lp)
```

```
### continue statement

l="Geeksfor Geeks"

for i in l:
  if (i=='e' or i=='s'):
    continue;
  else:
    print(i)
```

```
### continue statement

l="Geeksfor Geeks"

for i in l:
  if (i=='e' or i=='s'):
    break;
  else:
    print(i)
```

```
## pass statement

l="GeeksforGeeks"

for i in l:
  pass
print(i)
```
