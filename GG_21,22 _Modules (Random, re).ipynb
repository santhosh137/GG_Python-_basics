```
import re

p=re.compile('[a-e]')

print (p.findall("Aye"))
```

```
import re 

#\d is equivalent to [0-9]


p=re.compile('\d')

print (p.findall("I went ot him at 11 A.M on  4th july 1886"))


p=re.compile('\d+')

print (p.findall("I went ot him at 11 A.M on  4th july 1886"))
```

```
import re
### \w - equivalent to [a-zA-Z0-9]


p=re.compile('\w')

print (p.findall("He said * in some_lang"))

### \w+ - matches group of alpha numeric characters
p=re.compile('\w+')

print (p.findall("I went ot him at 11 A.M on  4th july 1886, he said **** in some $$$ language"))


#### \W - matches to non alphanumeric characters

p=re.compile('\W')

print (p.findall("he said **** in some $$$ language"))
```

```
### splitting astring using the split and re
```

```
from re import split

print (split('\W+','Do not worry, Sa will be home by 3 PM.'))

print (split('\d+','Do not worry, Sa will be home by 3 PM.'))
```

```
import re

print (re.sub('ub','`*', "Subject has uber booked already"))
print (re.sub('ub','`*', "Subject has uber booked already",count=1))

print (re.subn('ub','`*', "Subject has uber booked already"))
```
