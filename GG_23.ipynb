# Counters

Counters are  subclass of dictionary  i.e unordered collection of elements  and counts are stored in dictionary. It is also called as bags




```
### Intiakiation of counters in 3 ways

from collections import Counter

### with sequence f iterms
print (Counter (['B','B','A','C','A','B','B','A','C']))

### with dictionary
print (Counter({'A':3,'B':5,'C':2}))

##with keyword arguments
print (Counter(A=3,B=5,C=2))
```

```
### Updating a function

from collections import Counter
count=Counter()

print (count)

count.update([1,2,3,1,2,3,1,2,3])

print(count)

count.update([1,2,3])
print(count)
```

```
### Subtract function to find the difference between one coaunter  and another counter

from collections import Counter

c1=Counter (A=6,B=5,C=10)

c2=Counter(A=5,B=3,C=10)

c1.subtract(c2)

print (c1)
```
