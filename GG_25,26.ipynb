```
###Iteration over list

print ("list iterations")

l=["geeks","for","geeks"]

for i in l:
  print (i)

print ("\n Tuple iterations")

t=("geeks","for","geeks")

for i in t:
  print (i)


print ("\n String Iterations")

s="Geeks"

for i in s:
  print (i)


print ("dictionary iteration")

d=dict()

d["xyz"]=123
d["abc"]=345

for i in d:
  print ("%s %d"%(i,d[i]))
```
