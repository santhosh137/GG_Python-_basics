```
### iteration by for loop on string 

s="Hello world"

for i in s:
  print (i)

l=[i for i in s]
print (l)
```

```
##Iteration by  for loop on list

s=[1,2,3,34,4,5,6]

for i in s:
  print (i)

l=[i for i in s]
print (l)
```
