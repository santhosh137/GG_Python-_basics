```
###GG_6

### find() and rfind()
### used to find a position of substring in a string

## -1 -> string nnot found
### find  - . first occurence of the string

### rfind() - returns the last occurenece of the string


```

```
str ="This is a python tutorial"

print("The first occurence of 'is' is ar :")
print (str.find("is",0,len(str)))

print("The last occurence of 'is' is ar :")
print (str.rfind("is",0,len(str)))
```

```
## startswith and endswith()

```

```
A="GeeksforGeeks"
B="Geeks"

if (A.startswith(B)):
  print ("A starts with B")
else:
  print("A does not start with B")



if (A.endswith(B)):
  print ("A ends with B")
else:
  print("A does not endswith B")
```

```
### translation of string

## for  translation of a string we need a substituion of strings


## so first selecting the elements of string and substitute the selected elements 

str_="geeksforgeeks"

str1="gfo"
str2="abc"

mapped=str.maketrans(str1,str2)

#using translate () to translate using the mapping

print ("The string before translation  using mapped elements is :")
print (str_)

print("The string after translation using mapped elements is")

print (str_.translate(mapped))
```
